#  (1)استيراد المكتبات الأساسية


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder

# لضمان ظهور الرسوم داخل الدفتر
%matplotlib inline
# (2) تحديد مسارات الملفات كما في سؤالك
file_path_conflict = "/content/drive/MyDrive/data-minig/project/gdelt_conflict_1_0.csv"
file_path_fx       = "/content/drive/MyDrive/data-minig/project/Foreign_Exchange_Rates.csv"
```

# GDELT في df1  تحميل بيانات الصراعات

```python
# (3) تحميل ملف الصراعات في df1
df1 = pd.read_csv(file_path_conflict)

print("أول 5 صفوف من df1 قبل أي معالجة:")
print(df1.head(), "\n")
```

```python
print("أشكال الأعمدة (dtypes) في df1 قبل المعالجة:")
print(df1.dtypes, "\n")
```

```python
df1_clean = df1.copy()
```

```python
# (5) ترميز الأعمدة النصية في df1 إلى أرقام (Label Encoding)
categorical_cols_df1 = ["CountryCode", "CountryName", "EventRootDescr", "EventDescr"]
label_encoders = {}

for col in categorical_cols_df1:
    if col in df1_clean.columns:
        le = LabelEncoder()
        df1_clean[col] = le.fit_transform(df1_clean[col].astype(str))
        label_encoders[col] = le
    else:
        print(f"تحذير: العمود {col} غير موجود في df1، سيتم تجاهله في الترميز.")

print("أول 5 صفوف من df1 بعد ترميز الأعمدة النصية:")
print(df1_clean.head(), "\n")
```

```python
print("أشكال الأعمدة (dtypes) في df1 بعد الترميز:")
print(df1_clean.dtypes, "\n")
```

```python
print(df1_clean.describe())
```

```python
# (6) التأكد من أن كل الأعمدة رقمية
for col in df1_clean.columns:
    df1_clean[col] = pd.to_numeric(df1_clean[col], errors="coerce")
```

```python
# (7) معالجة القيم المفقودة في df1
df1_numeric = df1_clean.select_dtypes(include=["number"])
df1_numeric = df1_numeric.fillna(df1_numeric.mean())

print("شكل df1_numeric (الأعمدة الرقمية فقط):", df1_numeric.shape, "\n")
```

```python
# (8) حساب مصفوفة الارتباط لملف الصراعات df1
corr_df1 = df1_numeric.corr(method="pearson")

print("مصفوفة الارتباط لبيانات الصراعات (df1):")
print(corr_df1, "\n")
```

```python
# (9) رسم الخريطة الحرارية لمصفوفة الارتباط df1
plt.figure(figsize=(12, 8))
sns.heatmap(corr_df1, annot=False, cmap="coolwarm")
plt.title("Heatmap - Correlation Matrix (GDELT Conflict Data - df1)")
plt.tight_layout()
plt.show()
```

```python
# (10) تحميل ملف أسعار الصرف في df2
df2 = pd.read_csv(file_path_fx)

print("\nأول 5 صفوف من df2 قبل أي معالجة:")
print(df2.head(), "\n")
print("أشكال الأعمدة (dtypes) في df2 قبل المعالجة:")
print(df2.dtypes, "\n")
```

```python
# (11) إنشاء نسخة للعمل عليها
df2_clean = df2.copy()

# (12) حذف عمود Unnamed: 0 إذا كان موجوداً (عمود فهرس زائد)
if "Unnamed: 0" in df2_clean.columns:
    df2_clean = df2_clean.drop(columns=["Unnamed: 0"])

# (13) تحويل عمود Time Serie إلى datetime (اختياري لكن مفيد)
if "Time Serie" in df2_clean.columns:
    df2_clean["Time Serie"] = pd.to_datetime(df2_clean["Time Serie"], errors="coerce")

# (14) تحويل كل الأعمدة (ما عدا Time Serie) إلى أرقام
for col in df2_clean.columns:
    if col == "Time Serie":
        continue  # نتركه كتاريخ
    df2_clean[col] = pd.to_numeric(df2_clean[col], errors="coerce")
```

```python
print("أول 5 صفوف من df2 بعد تنظيف الأنواع:")
print(df2_clean.head(), "\n")
print("أشكال الأعمدة (dtypes) في df2 بعد التحويل:")
print(df2_clean.dtypes, "\n")
```

```python
# (16) معالجة القيم المفقودة في df2 عن طريق تعويضها بمتوسطات الأعمدة
df2_numeric = df2_numeric.fillna(df2_numeric.mean())

print("شكل df2_numeric (الأعمدة الرقمية فقط):", df2_numeric.shape, "\n")

# (17) حساب مصفوفة الارتباط لملف أسعار الصرف df2
corr_df2 = df2_numeric.corr(method="pearson")
```

```python

print("مصفوفة الارتباط لبيانات أسعار الصرف (df2):")
print(corr_df2, "\n")

```

```python
# (18) رسم الخريطة الحرارية لمصفوفة الارتباط df2
plt.figure(figsize=(14, 10))
sns.heatmap(corr_df2, annot=False, cmap="coolwarm")
plt.title("Heatmap - Correlation Matrix (Foreign Exchange Rates - df2)")
plt.tight_layout()
plt.show()
```
